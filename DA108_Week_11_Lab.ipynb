```python
# Problem 1: Calculate the inverse of a matrix using NumPy
import numpy as np

matrix = np.array([[1, 3],
                   [2, 4]])
determinant = np.linalg.det(matrix)

if determinant != 0:
    inverse = np.linalg.inv(matrix)
    print("Matrix:\n", matrix)
    print("Inverse:\n", inverse)

else:
    print("The given matrix is singular.")
```
